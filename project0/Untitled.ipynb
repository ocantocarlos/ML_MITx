```python
import numpy as np
import matplotlib.pyplot as plt
def check(point, theta, label,i=0):
    sign = (label*np.dot(theta,point))
    print ("y({})(x({}).theta) =".format(str(i),str(i)),sign)
    if (sign)<=0:
        return True
    return False

def update_theta(point, theta, label):
    return theta + point*label

def perceptron(points,labels, theta = np.array([0,0]), iterations=1):
    errors = 0
    for it in range(iterations):
        print("\n----------------iteration {}----------------".format(it+1))
        for i, point in enumerate(points):
            print("\npoint",point,i)
            print("theta",theta)
            print("y",labels[i])
            check_ = check(np.array(point), np.array(theta), labels[i],i=i)
            
            if check_:
                errors+=1
                theta = update_theta(np.array(point), np.array(theta), labels[i])
                
                print("\n***********new theta",theta)
                
            else:
                print("theta not updated")
                
points = [[-1,-1],[1,0],[-1,1.5]]
labels = [1, -1, 1]
perceptron(points,labels, iterations=2)         
```
